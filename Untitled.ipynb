<h3>Q4<br> a. Classify the Iris flower dataset using Support Vector Machine (SVM) and identify the
flower type based on given input data.<br>
b. Predict whether breast cancer is benign or malignant using the K-Nearest Neighbors
(KNN) Classifier for a given dataset</h3>

```python
from sklearn.svm import SVC
from sklearn.datasets import *
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
iris_df['species'] = iris_df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
x_train_4a, x_test_4a, y_train_4a, y_test_4a = tts(iris_df.iloc[:, :-1], iris_df.iloc[:, -1], test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
x_train_4a = scaler.fit_transform(x_train_4a)
x_test_4a = scaler.transform(x_test_4a)
```

```python
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(x_train_4a, y_train_4a)
```

```python
y_pred6 = svm_classifier.predict(x_test_4a)
accuracy = accuracy_score(y_test_4a, y_pred6)
print(f"Accuracy: {accuracy * 100:.2f}%")
```
